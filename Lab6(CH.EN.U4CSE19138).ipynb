```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import nltk
from collections import Counter
text = open('tomsawyer.txt','r',encoding='utf8')
raw = text.read()
sentence = nltk.word_tokenize(raw)
pos = nltk.pos_tag(sentence)
print(pos)
the_count = Counter(tag for _, tag in pos)
print(the_count)
```

```python
pos = np.arange(len(the_count.keys()))
width = 1.0
ax = plt.axes()
ax.set_xticks(pos + (width / 2))
ax.set_xticklabels(the_count.keys())
plt.bar(the_count.keys(), the_count.values(), width, color='g')
```
